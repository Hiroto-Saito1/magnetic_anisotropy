```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

mae_conv = pd.read_table("mae_conv.txt")

plt.figure()
plt.plot(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3)
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3)
plt.grid()
plt.xlabel("$N_k$")
plt.ylabel("$\Delta E$ [meV]")
plt.savefig("mae_conv.pdf")
plt.show()
```
