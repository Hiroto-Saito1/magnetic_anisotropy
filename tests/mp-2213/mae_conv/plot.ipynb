```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import toml

with open("../result.toml") as f:
    result = toml.load(f)
err = float(result["M_diff_ave [meV]"])
mae_conv = pd.read_table("./mae_conv.txt")

plt.figure()
plt.rcParams["font.size"] = 16
plt.rcParams['figure.subplot.left'] = 0.20
plt.rcParams['figure.subplot.bottom'] = 0.20
plt.errorbar(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**6, err/2*np.ones(len(mae_conv))*10**3, ecolor="green")
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**6)
plt.grid()
plt.xticks(ticks=[0,50,100,150,200,250,300,350],
           labels=["$0$","$50^3$","$100^3$","$150^3$","$200^3$","$250^3$","$300^3$","$350^3$"])
plt.xlabel("# of $k$ points")
plt.ylabel("$E[100] - E[001]$ [$\mu$eV]")
plt.xlim(0,300)

plt.axes([0.55, 0.67, 0.35, 0.3])
plt.errorbar(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**6, err/2*np.ones(len(mae_conv))*10**3, ecolor="green")
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**6)
plt.grid()
plt.xlim(150, 300)
plt.xticks(ticks=[150,200,250,300], labels=["$150^3$","$200^3$","$250^3$","$300^3$"])
plt.ylim(80,90)
plt.savefig("mae_conv.pdf")
plt.show()
```

```python
plt.figure()
plt.errorbar(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3, err/2*np.ones(len(mae_conv)), ecolor="green")
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3)
plt.grid()
plt.xlabel("$N_k$")
plt.ylabel("$\Delta E$ [meV]")
plt.xlim(60,)
plt.ylim(0.08,0.09)
plt.show()
```
