```python
import numpy as np 
import matplotlib.pylab  as plt
from pathlib import Path
import sys
import os

src_path = os.path.abspath(os.path.join('..', '..', '..', 'src'))
sys.path.append(src_path)
from mag_rotation import MagRotation
```

```python
tb_dat = Path("../pwscf_rel_sym/mae/wan/pwscf_py_tb.dat")
ham = MagRotation(tb_dat=tb_dat, extract_only_x_component=False)
```

```python
hopping = ham.spinor_to_vector(ham.reshape_hrs_to_ham_spinor(ham.hopping)[ham.ir0])
mag = ham.spinor_to_vector(ham.reshape_hrs_to_ham_spinor(ham.mag_orig)[ham.ir0])
soc = ham.spinor_to_vector(ham.reshape_hrs_to_ham_spinor(ham.soc)[ham.ir0])
```

```python
fontsize = 14
plt.imshow(hopping[0].real)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=fontsize)
cbar.set_label("eV", fontsize=fontsize)

ticks = [0,1,4, 9,10,13]
labels = ["$4s$","$4p$","$3d$", "$4s$","$4p$","$3d$"]
plt.xticks(ticks, labels, fontsize=12)
plt.text(0, 20, 'Fe', ha='center', va='center', fontsize=fontsize)
plt.text(9, 20, 'Ni', ha='center', va='center', fontsize=fontsize)
plt.yticks(ticks, labels, fontsize=12)
plt.text(-3, 0, 'Fe', ha='center', va='center', fontsize=fontsize)
plt.text(-3, 9, 'Ni', ha='center', va='center', fontsize=fontsize)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)
plt.savefig("t.pdf")
plt.show()
```
