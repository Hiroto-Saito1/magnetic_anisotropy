```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
```

```python
data_calc = pd.read_table("./bandfilling.txt")
data_orig = pd.read_table("./bandfilling_orig.txt")

num_valence = 18
L = 1000
band_filling = np.linspace(num_valence - 2, num_valence + 2, L)
plt.rcParams["font.size"] = 16
plt.rcParams['figure.subplot.left'] = 0.20
plt.rcParams['figure.subplot.bottom'] = 0.20 
plt.figure()
plt.plot(band_filling, data_orig.iloc[:, 0] * 10 ** 6, label="TB", color="black")
plt.plot(band_filling, data_calc.iloc[:, 0] * 10 ** 6, label="TB-EM", linestyle="dashed", color="black")
plt.plot(band_filling, data_calc.iloc[:, 1] * 10 ** 6, label="TB-M", linestyle="dotted", color="black")
plt.grid()
plt.axvline(x=num_valence, color="grey")
plt.legend()
plt.xlabel("Number of electrons")
plt.ylabel("$E[100] - E[001]$ $[\mathrm{\mu eV}/f.u.]$")
plt.savefig("./bandfilling.pdf")
```
