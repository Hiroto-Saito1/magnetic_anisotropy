```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import toml

with open("../result.toml") as f:
    result = toml.load(f)
err = float(result["M_diff_ave [meV]"])
mae_conv = pd.read_table("./mae_conv.txt")

plt.figure()
plt.errorbar(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3, err/2*np.ones(len(mae_conv)), ecolor="green")
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3)
plt.grid()
plt.xticks(ticks=[0,50,100,150,200,250,300,350],
           labels=["$0$","$50^3$","$100^3$","$150^3$","$200^3$","$250^3$","$300^3$","$350^3$"])
plt.xlabel("Number of $k$ points")
plt.ylabel("$E[100] - E[001]$ [meV/f.u.]")
plt.xlim(0, 300)

plt.axes([0.512, 0.615, 0.39, 0.3])
plt.errorbar(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3, err/2*np.ones(len(mae_conv)), ecolor="green")
plt.scatter(mae_conv.iloc[:,0], mae_conv.iloc[:,1]*10**3)
plt.grid()
plt.xlim(150, 300)
plt.xticks(ticks=[150,200,250,300], labels=["$150^3$","$200^3$","$250^3$","$300^3$"])
plt.ylim(2.37, 2.405)
plt.savefig("mae_conv.pdf")
plt.show()
```
