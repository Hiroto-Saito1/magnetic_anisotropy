```python
import numpy as np
import matplotlib.pylab as plt
```

```python
energy = np.zeros([45, 45])
f = open("angle_dep.txt", "r")
f.readline()
f.readline()
for t_i in range(45):
    for p_j in range(45):
        energy[t_i, p_j] = float(f.readline())
f.close()
```

```python
plt.figure()
plt.rcParams["font.size"] = 16
plt.rcParams['figure.subplot.left'] = 0.20
plt.rcParams['figure.subplot.bottom'] = 0.20    
phi = np.linspace(0, 180, 45)
theta = np.linspace(0, 180, 45)
plt.figure()
plt.pcolor(
    phi, theta, (energy - energy[0, 0]) * 10 ** 3, shading="auto"
)
plt.xlabel("$\phi$ [deg]")
plt.ylabel("$\\theta$ [deg]")
plt.colorbar(label="$dE$ [meV]")
plt.xticks([0, 45, 90, 135, 180], ["0", "45", "90", "135", "180"])
plt.yticks([0, 45, 90, 135, 180], ["0", "45", "90", "135", "180"])
plt.grid()
plt.savefig("angle_dep.pdf")
plt.show()
```
