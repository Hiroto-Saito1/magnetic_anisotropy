```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
```

```python
data_calc = pd.read_table("./bandfilling.txt")
data_orig = pd.read_table("./bandfilling_orig.txt")

num_valence = 18
L = 1000
band_filling = np.linspace(num_valence - 2, num_valence + 2, L)
plt.plot(band_filling, data_orig.iloc[:, 0] * 10 ** 3, label="original", color="black")
plt.plot(band_filling, data_calc.iloc[:, 0] * 10 ** 3, label="with extraction", linestyle="dashed", color="black")
plt.plot(band_filling, data_calc.iloc[:, 1] * 10 ** 3, label="w/o extraction", linestyle="dotted", color="black")
plt.grid()
plt.axvline(x=num_valence, color="grey")
plt.legend()
plt.xlabel("# of electrons")
plt.ylabel("$E[100] - E[001]$ $[\mathrm{meV}/f.u.]$")
plt.savefig("./bandfilling.pdf")
```
