```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
```

```python
data_calc = pd.read_table("./bandfilling.txt")
data_orig = pd.read_table("./bandfilling_orig.txt")

num_valence = 18
L = 1000
band_filling = np.linspace(num_valence - 2, num_valence + 2, L)
plt.rcParams["font.size"] = 16
plt.rcParams['figure.subplot.left'] = 0.20
plt.rcParams['figure.subplot.bottom'] = 0.20 
plt.figure()
plt.plot(band_filling, data_orig.iloc[:, 0] * 10 ** 3, label="DFT", color="black")
plt.plot(band_filling, data_calc.iloc[:, 0] * 10 ** 3, label="TB-EM", linestyle="dashed", color="black")
plt.plot(band_filling, data_calc.iloc[:, 1] * 10 ** 3, label="TB-OM", linestyle="dotted", color="black")
plt.grid()
plt.xlim(16, 20)
plt.axvline(x=num_valence, color="grey")
plt.legend()
plt.xlabel("Number of electrons")
plt.ylabel("$E_{[100]} - E_{[001]}$ $[\mathrm{meV}/f.u.]$")
plt.savefig("./bandfilling.pdf")
```

```python
# TB-EM  DFT

num_valence = 18
L = 1000
band_filling = np.linspace(num_valence - 2, num_valence + 2, L)
plt.rcParams["font.size"] = 16
plt.rcParams['figure.subplot.left'] = 0.20
plt.rcParams['figure.subplot.bottom'] = 0.20 
diff = np.abs(data_orig.iloc[:, 0] - data_calc.iloc[:, 0]) * 10 ** 6
plt.figure()
plt.plot(band_filling, diff, label="|DFT - TB-EM|", color="black")
plt.grid()
plt.xlim(16, 20)
plt.axvline(x=num_valence, color="grey")
plt.legend()
plt.xlabel("Number of electrons")

print(f"Average diff: {np.mean(diff)} [ueV/f.u.]")
```
